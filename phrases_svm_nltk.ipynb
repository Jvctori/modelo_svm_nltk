```python
# Criação dos DataFrames para representar diferentes classes de sentimentos, atribuindo valores numéricos a cada categoria de frases.
import pandas as pd
dfOrgulho = pd.DataFrame({'phase': pride_phrases,
                          'class': 0})
dfAmor = pd.DataFrame({'phase': love_phrases,
                          'class': 1})
dfTristeza = pd.DataFrame({'phase': sadness_phrases,
                          'class': 2})
```

```python
dfOrgulho.head()
```

```python
dfOrgulho.shape
```

```python
dfAmor.shape
```

```python
dfTristeza.shape
```

```python
# Concatenação dos Data Frames
df_all = pd.concat([dfOrgulho,dfAmor,dfTristeza],axis=0,ignore_index=True)
```

```python
df_all.head()
```

```python
df_all.shape
```

```python
# Importação do vader
# Criação das colunas com os valores positivos, negativos, neutros e compound (métrica de pontuação geral)
# Valores próximos a 1 indicam um sentimento extremamente positivo.
# Valores próximos a -1 indicam um sentimento extremamente negativo.
# Valores próximos a 0 indicam neutralidade ou ausência de um sentimento polarizado.
from nltk.sentiment import vader
sia = vader.SentimentIntensityAnalyzer()
df_all['pos'] = df_all['phase'].apply(lambda x: sia.polarity_scores(x)['pos'])
df_all['neg'] = df_all['phase'].apply(lambda x: sia.polarity_scores(x)['neg'])
df_all['neu'] = df_all['phase'].apply(lambda x: sia.polarity_scores(x)['neu'])
df_all['compound'] = df_all['phase'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
df_all.dtypes
```

```python
# Importação do metodo de separaçãod e treino e teste
from sklearn.model_selection import train_test_split
import numpy as np

# Atribuindo a X um array numpy com os valores relevantes para a classificação 
X = df_all[['pos','neg','neu','compound']].values
# Atribuindo a y um array numpy com os valores que representam as classes (sentimentos)
y = df_all['class'].values
# Separando em base de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Importando o modelo Super Vector Machine do sci-kit learn
from sklearn import svm
modelSVC = svm.SVC()
# Passando os dados de treino para o Modelo
modelSVC.fit(X_train,y_train.ravel())
```

```python
# Usando a função para prever as classes dos dados de teste
predictionSVC = modelSVC.predict(X_test)
```

```python
# Importação das métricas de avaliação do modelo
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um heatmap para visualizar a matriz de confusão
conf_matrix = confusion_matrix(y_test, predictionSVC)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Relatorio de Classificação
print(classification_report(y_test, predictionSVC))
```

```python
# Acurácia
accuracy = accuracy_score(y_test, predictionSVC)
print(f'Acurácia do modelo SVM: {accuracy:.2f}')
```
